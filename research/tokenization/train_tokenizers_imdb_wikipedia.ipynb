```python
import pandas as pd
from sklearn.model_selection import train_test_split
import os
```

```python
from datasets import load_dataset, concatenate_datasets

ds = load_dataset("imdb")

merged = concatenate_datasets([ds["test"], ds["train"]])
df = pd.DataFrame(merged)
df["text"] = df["text"].str.replace(r"<br\s*/?>", "", regex=True)
train_df, test_df = train_test_split(df, test_size=0.2, random_state=42, stratify=df['label'])
```

```python
with open('./data/raw/imdb.txt', 'w', encoding='utf-8') as f:
    for review in train_df['text'].astype(str):
        f.write(review + '\n')
```

```python
def file_length(path: str) -> int:
    with open(path, "r", encoding="utf-8") as f:
        content = f.read()
    return len(content)

def copy_first_n_chars(input_path: str, n: int, output_path: str):
    with open(input_path, "r", encoding="utf-8") as f:
        content = f.read()

    to_write = content[:n]

    with open(output_path, "w", encoding="utf-8") as out:
        out.write(to_write)
```

```python
proportion = 10
```

```python
n = file_length('./data/raw/imdb.txt')
```
