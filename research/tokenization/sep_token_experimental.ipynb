```python
import torch
from src.textclf_transformer import *
```

```python
tok = WordPieceTokenizerWrapper()
path = Path("src/textclf_transformer/tokenizer/BERT_original")
tok.load(tokenizer_dir=path)
tok.tokenizer.sep_token_id
```

```python
tok.tokenizer.cls_token_id
```

```python
datasets = ['hyperpartisan_test','hyperpartisan_train','hyperpartisan_val',
            'arxiv_train','arxiv_test','arxiv_val']


for d in datasets:
    ds = torch.load(f'data/tokenized/{d}.pt', weights_only=False)
    ds[:][0][:, 512::512] = tok.tokenizer.cls_token_id
    ds[:][0][ds[:][1]] = 0 # mask pad tokens

    save_dir = Path("data/tokenized")
    torch.save(ds, save_dir / f"{d}_cls.pt")
```
